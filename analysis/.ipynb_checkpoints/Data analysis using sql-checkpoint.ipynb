### 13. Calculate the percentage distribution of bookings by booking status.

```python

try:
    with mysql.connector.connect(**connection) as conn:
        with conn.cursor() as cur:
            query="""
                    select 
                        booking_status,
                        Round(count(*)*100.0/(select count(*) from bookings),2) as Percentage
                    from
                        bookings
                    group by
                        booking_status;
                  """
            cur.execute(query)
            data=cur.fetchall()
            column=[col[0] for col in cur.description]
            data=pd.DataFrame(data,columns=column)
            print("="*20,"Calculate the percentage distribution of bookings by booking status.","="*20)
            print(data)
            print("="*110)
except Exception as e:
    print("Database error:",e)
finally:
    print("ALL DONE")
```

### 14. Compare total revenue generated on weekends vs weekdays.

```python

try:
    with mysql.connector.connect(**connection) as conn:
        with conn.cursor() as cur:
            query="""
                     select 
                         case
                         when
                             DAYOFWEEK(Date) IN (1,7) THEN 'Weekday'
                         else
                             'Weekend'
                         END AS 
                             day_type,
                        sum(booking_value) as Total_Revenue_Generated
                    from 
                        bookings
                    where
                        booking_status='successfull'
                    group by day_type;
                  """
            cur.execute(query)
            data=cur.fetchall()
            column=[col[0] for col in cur.description]
            data=pd.DataFrame(data,columns=column)
            print("="*30,"total revenue generated on weekends vs weekdays." ,"="*30)
            print(data)
            print("="*110)
except Exception as e:
    print("Database error:",e)
finally:
    print("ALL DONE")
```

### 17. Identify the most common reasons for driver cancellations.

```python
try:
    with mysql.connector.connect(**connection) as conn:
        with conn.cursor() as cur:
            query="""
                    select 
                        Driver_Cancel_Reason,
                        count(Driver_Cancel_Reason) as occurence_of_reason
                    from
                        bookings
                    group by Driver_Cancel_Reason
                    order by occurence_of_reason desc;
                    """
            cur.execute(query)
            data=cur.fetchall()
            column=[col[0] for col in cur.description]
            data=pd.DataFrame(data,columns=column)
            print("="*30,"the average driver rating by vehicle type for completed rides","="*30)
            print(data)
            common=data.iloc[0]
            print(f"most common reasons for driver cancellations: {common['Driver_Cancel_Reason']} \n occured {common['occurence_of_reason']} times")
except Exception as e:
    print("Database error:",e)
finally:
    print("ALL DONE")
```

### 19. Determine which vehicle type has the highest average ride distance.


```python
# 19. Determine which vehicle type has the highest average ride distance.

try:
    with mysql.connector.connect(**connection) as conn:
        with conn.cursor() as cur:
            query = """
                SELECT 
                   Vehicle_Type ,Round(AVG(Ride_Distance),2) AS Average_Ride_Distance 
                FROM 
                    bookings
                 Group by Vehicle_Type
                 order by Average_Ride_Distance  desc;
            """
            cur.execute(query)

            data = cur.fetchall()
            columns = [col[0] for col in cur.description]

            data = pd.DataFrame(data, columns=columns)
            print("="*20,"the average ride distance for each vehicle type","="*20)
            print(data)
            highest=data.iloc[0]
            print(f" {highest['Vehicle_Type']} has the highest average ride distance of {highest['Average_Ride_Distance']} ")
            print("="*90)

except Exception as e:
    print("Database error:", e)

finally:
    print("Its running")
```

### 20. Find customers with repeated cancellations (customer behavior risk analysis).

```python
# Find customers with repeated cancellations (customer behavior risk analysis)
try:
    with mysql.connector.connect(**connection) as conn:
        with conn.cursor() as cur:
            query="""
                    SELECT
                        Customer_ID,
                        COUNT(*) AS cancellation_count
                    FROM 
                        bookings
                    WHERE 
                        Cancelled_Rides_by_Customer = 1
                    GROUP BY 
                        Customer_ID
                    HAVING 
                        COUNT(*) > 1
                    ORDER BY 
                        cancellation_count DESC;
                    """
            cur.execute(query)
            data=cur.fetchall()
            column=[col[0] for col in cur.description]
            data=pd.DataFrame(data,columns=column)
            print("="*30,"the most common reasons for customer cancellations.","="*30)
            print(data)
            print("="*112)
except Exception as e:
    print("Database error:",e)
finally:
    print("ALL DONE")
```
